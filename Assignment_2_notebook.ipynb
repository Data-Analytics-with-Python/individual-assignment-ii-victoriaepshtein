## Task A (30 points)

In this question, we investigate how customers' online activity relates to their web purchasing behavior.

### Task A.1 (15 points)

i. Choose an appropriate chart type that can reveal **whether people who visit the website more often also make more purchases online**. In your chart, each data record corresponding to a customer should be represented by a **point**, the coordinates of which represent the *number of website visits last month* and the *number of online purchases*, respectively.
  * Develop the chart with visual encodings --  use **size** to encode customers' **income** and **color** to encode the **campaign response** (whether accepted any campaigns) in your chart.
  * When a user hovers their mouse on elements in the chart, let the chart show relevant information on *income*, *education*, and *marital status* of the corresponding data record.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
  * Do more frequent site visitors tend to buy more?
  * How does income affect online purchases?
  * Are campaign responders visually distinguishable? If so, how are they different?


```python
import plotly.express as px

# Drop rows with missing values as stated in the initial instructions (assuming df_cleaned already handles this, but adding safeguard)
df_cleaned = df.dropna().copy()

# Create 'CampaignResponse' column: 1 if any campaign was accepted, 0 otherwise
df_cleaned['CampaignResponse'] = df_cleaned[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].any(axis=1).astype(int)

# Convert CampaignResponse to string for better categorical coloring in Plotly
df_cleaned['CampaignResponse'] = df_cleaned['CampaignResponse'].map({0: 'No', 1: 'Yes'})

# Create the scatter plot
fig = px.scatter(
    df_cleaned,
    x='NumWebVisitsMonth',
    y='NumWebPurchases',
    size='Income',
    color='CampaignResponse',
    hover_name='ID',
    hover_data=['Income', 'Education', 'Marital_Status', 'NumWebVisitsMonth', 'NumWebPurchases'],
    title='Online Purchases vs. Website Visits by Income and Campaign Response',
    labels={
        'NumWebVisitsMonth': 'Number of Website Visits Last Month',
        'NumWebPurchases': 'Number of Online Purchases',
        'Income': 'Income',
        'CampaignResponse': 'Accepted Campaign'
    },
    size_max=25 # Adjust size_max to make size differences easier to see
)

fig.show()
```



*   **Do more frequent site visitors tend to buy more?**

*No, more frequent site visitors did not necessarily buy more.*

*  **How does income affect online purchases?**

*People with higher income tended to buy something regardless of whether they visited the wabsite, whereas those with lower income tended to visit the website, sometimes even frequently, and not buy anything.*

*   **Are campaign responders visually distinguishable? If so, how are they different?**

*Yes, they are visually distinguishable by colour - red vs blue. Those that accepted the campaign were more likely to make a higher number of online purchases with less website visits.*

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
import plotly.express as px
import pandas as pd

# Re-define df_cleaned to ensure it's available in this cell's scope
df = pd.read_csv('/content/natadata.csv')
df_cleaned = df.dropna().copy()

# Create 'CampaignResponse' column if it's needed for other parts of the notebook
# (though not directly used in B.1, it's good practice to keep the df consistent if this cell is run independently)
if 'CampaignResponse' not in df_cleaned.columns:
    df_cleaned['CampaignResponse'] = df_cleaned[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].any(axis=1).astype(int)
    df_cleaned['CampaignResponse'] = df_cleaned['CampaignResponse'].map({0: 'No', 1: 'Yes'})

# Define product columns for melting
product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Melt the DataFrame to transform product spending columns into a long format
df_melted = df_cleaned.melt(id_vars=['Education'], value_vars=product_columns, var_name='ProductCategory', value_name='Spending')

# Group by Education and ProductCategory and sum the spending
df_grouped = df_melted.groupby(['Education', 'ProductCategory'], as_index=False)['Spending'].sum()

# Create the stacked bar chart
fig = px.bar(
    df_grouped,
    x='Education',
    y='Spending',
    color='ProductCategory',
    title='Total Spending by Education Level and Product Category',
    labels={
        'Education': 'Education Level',
        'Spending': 'Total Spending',
        'ProductCategory': 'Product Category'
    },
    barmode='stack'
)

# Remove border lines as per the hint
fig.update_traces(opacity=1, marker_line_width=0)

fig.show()
```



*  **Which customer group contributes the highest revenue?**

*Graduation contributed to the highest revenue.*

## Part C (30 points)

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
import plotly.express as px
import pandas as pd

# Calculate Age (assuming current year is 2024 for generation calculation)
df_cleaned['Age'] = 2024 - df_cleaned['Year_Birth']

# Define a function to assign generations
def assign_generation(row):
    if 1981 <= row['Year_Birth'] <= 1996: # Millennials (born 1981-1996)
        return 'Millennials'
    elif 1965 <= row['Year_Birth'] <= 1980: # Gen X (born 1965-1980)
        return 'Gen X'
    else:
        return 'Other'

# Apply the function to create the 'Generation' column
df_cleaned['Generation'] = df_cleaned.apply(assign_generation, axis=1)

# Filter for Gen X and Millennials
df_generations = df_cleaned[df_cleaned['Generation'].isin(['Gen X', 'Millennials'])].copy()

# Convert Kidhome to string for categorical display
df_generations['Kidhome_Category'] = df_generations['Kidhome'].astype(str)

# Create the violin plot
fig = px.violin(
    df_generations,
    x='Generation',
    y='NumWebPurchases',
    color='Kidhome_Category', # Color by number of kids at home
    box=True, # Show quartiles and median within the violin
    points="outliers", # Show outliers
    title='Distribution of Online Purchases by Generation and Number of Kids at Home',
    labels={
        'Generation': 'Customer Generation',
        'NumWebPurchases': 'Number of Online Purchases',
        'Kidhome_Category': 'Number of Kids at Home'
    },
    category_orders={'Generation': ['Gen X', 'Millennials'], 'Kidhome_Category': ['0', '1', '2']}
)

fig.show()
```



*   **How would you comment on the pattern of online purchases across different generations (Gen X vs. Millennials)?**

*I would say that the number of online purchases is actually relatively well distributed across Gen X and Millenials.*

*   **How would you comment on the pattern of online purchases across the different subgroups of customers based on the number of kids at home?**

*Individuals without kids in both Gen X and Millenial categories make the most online purchases*
